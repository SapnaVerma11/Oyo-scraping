```python
# importing the necessary libraries
import selenium 
import time
# Importing required Exceptions which needs to handled
from selenium.common.exceptions import NoSuchElementException,StaleElementReferenceException

PATH = "C:\Program Files (x86)\chromedriver.exe"
from selenium import webdriver 

# Activating the chrome browser
driver = webdriver.Chrome(PATH)

# Opening the homepage of OYO 
driver.get("https://www.oyorooms.com/")
```

```python
# scraping data for noida city 
search_bar = driver.find_element_by_xpath('/html/body/div[2]/div/div[3]/div[1]/div[3]/div/div/div/div[1]/div/div/form/input')
search_bar.send_keys('Noida')
time.sleep(3)

#clicking on noida from drop down list 
noida_btn = driver.find_element_by_xpath('/html/body/div[2]/div/div[3]/div[1]/div[3]/div/div/div/div[1]/div/span/div/div/div[1]/div/div[2]')
noida_btn.click()
time.sleep(2)
```

```python
# clicking on search button 
search_btn = driver.find_element_by_xpath('/html/body/div[2]/div/div[3]/div[1]/div[3]/div/div[1]/div/div[4]/button')
search_btn.click()
```

```python
# collecting urls 
urls = []
# clicking on load more button(found 20 more results button)
for i in range(0,3):
    driver.find_element_by_xpath('//span[@class="loadMoreIcon"]').click()
try:    
    for i in driver.find_elements_by_xpath('//a[@class="c-nn640c u-width100"]'):
        urls.append(i.get_attribute('href'))
except:
    pass
```

```python
#scraping the titles
titles = []
try:
    for i in driver.find_elements_by_xpath('//a[@class="c-nn640c u-width100"]/h3'):
        titles.append(i.text)
except:
    titles.append('No Title')
                  

#scraping the locations of rooms
location = []
try:
    for i in driver.find_elements_by_xpath('//span[@class="u-line--clamp-2"]'):
        location.append(i.text)
except:
    location.append('No location')
                  
                  
#scraping the price of rooms 
price = []
try:  
    for i in driver.find_elements_by_xpath('//span[@class="listingPrice__finalPrice"]'):
        price.append(i.text)
except:
    price.append('No price')   
                  
#scraping the discount of rooms 
discount = []
try:
    for i in driver.find_elements_by_xpath('//span[@class="listingPrice__percentage"]'):
        discount.append(i.text)
except :
       discount.append('No discount')
                      
```

```python
print(len(titles),len(location),len(price),len(discount))
```

```python
import pandas as pd 
df = pd.DataFrame({'Titles':titles,'Location':location,'Price':price,'Discount':discount})
df
```
